# 2.2 Manipulating the data

In order to manipulate the data, we will remove the duplicate rows, remove the missing data, and remove all outliers data.

### 2.2.1 Dropping the duplicate rows

in order to drop all the duplicate rows, we used the drop duplicates function. Also, the parameter of inplace is set to true, which indicated that the old data DataFrame will be updated to the new one that not contain duplicate rows. ### 2.2.2 Filtering Missing Values

Pandas uses the value of NaN to represent missing values. In order to tell whether or not a value is missing in a DataFrame, isnull() function is used to filter the rows with missing values.


```python
app[app['App'].isnull()]
```

The output of the execution indicates that the app column does not contain any missing value.

```python
app[app['Category'].isnull()]
```

The output of the execution indicates that the Category column does not contain any missing value.

```python
app[app['Installs'].isnull()]
```

The output of the execution indicates that the Installs column does not contain any missing value.

```python
app[app['Price'].isnull()]
```

The output of the execution indicates that the Price column does not contain any missing value.

As mentioned before, the Rating column is the only column that contains Missing Values. In order to manipulate those Missing Values, we decided to take the mean of the Rating column and replace all the Missing Values in that column with that mean value. To do that, we used the Pandas mean function, since it excludes missing values representing missing data, and therefore does not affecting the result of calculating the mean value.


```python
print ('Pandas mean function:', app['Rating'].mean())
```

The output of the execution indicates that the mean value of Rating column is 4.16, hence, the fillna function is used to replace all the missing values with the mean value. 

```python
app = app.fillna(value={'Rating': app['Rating'].mean()})

app
```

To make sure that all the missing values of Rating column are replaced, we again called the isnull function for Rating column

```python
app[app['Rating'].isnull()]
```

The output of the execution indicates that the Rating column does not contains any missing values now.

#### 2.2.3 Filtering the outliers values

 we called the sort function for Size column in order to detect all the outliers that have 0 value.


```python
app.sort_values(by='Size', ascending=False, inplace=True)
app.tail(30)
```

The output of the execution indicates that the Installs column contains outliers values.To manipulated those outliers values, we find the mean for Size column and then the loc function is used to replace all the outlier values.





```python
print ('Pandas mean function:', app['Size'].mean())
```

The output of the execution indicates that the mean value of Size column is 2.39 M, hence, the loc function is used to replace all the outlier values with that mean value. 

```python
app.loc[app['Size'] == 0.00, ['Size']] = 20.39
app
```

To make sure that all the outlier values of Size column are replaced, we again called the sort function for Szie column.

```python
app.sort_values(by='Size', ascending=False, inplace=True)
app.tail(30)
```

 The output of the execution indicates that the Size column does not contains any outliers values now.
